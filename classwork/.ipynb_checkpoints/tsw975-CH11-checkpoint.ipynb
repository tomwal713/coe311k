```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Define known data points
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 0, 1, 0, 1])

# Cubic spline interpolation
cs = CubicSpline(x, y, bc_type='natural')

# Polynomial fit
# degree=3 gives a cubic polynomial fit
coeffs = np.polyfit(x, y, 3)
poly = np.poly1d(coeffs)

# Smooth x-values for plotting
x_interp = np.linspace(0, 5, 200)

# Evaluate both curves
y_spline = cs(x_interp)
y_poly = poly(x_interp)
```

```python
# Plot everything on the same graph
plt.figure(figsize=(8, 5))
plt.plot(x, y, 'o', label='Data points')
plt.plot(x_interp, y_spline, label='Cubic Spline')
plt.plot(x_interp, y_poly, label='Polynomial Fit (degree 3)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Spline vs Polynomial Fit')
plt.legend()
plt.grid(True)
plt.show()
```
