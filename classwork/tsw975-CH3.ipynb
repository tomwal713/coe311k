```python
import random
import numpy as np #for true value of pi
import matplotlib.pyplot as plt
pi_true = np.pi #storing true value of pi
```

```python
def monte_carlo_pi(n): #function using to approximate pi
    inside = 0 #counts how many points will fall inside the quarter circle

    for _ in range(n): #here I'm generating points inside the unit square, generating n points
        x = random.random()
        y = random.random()

        if x*x + y*y <= 1: #checking if the points lie within the circle
            inside += 1 #if the point is inside the circle, +1

    return 4 * inside / n #this estimates the area of the quarter circle, multiply by 4 to get pi
```

```python
num_runs = 100
samples = 1000

absolute_errors = [] #storing both errors
relative_errors = []

for _ in range(num_runs):
    pi_estimate = monte_carlo_pi(samples)

    abs_error = abs(pi_true - pi_estimate) #Absolute error: |x_true - x_approx|

    rel_error = abs((pi_true - pi_estimate) / pi_true) #Relative error: |(x_true - x_approx) / x_true|

    absolute_errors.append(abs_error) #each loop will append the error found, to be plotted
    relative_errors.append(rel_error)
```

```python
plt.figure()

plt.plot(range(1, num_runs + 1), absolute_errors, 
         label="Absolute Error", marker='o') #plotting absolute error on blue line with o markers

plt.plot(range(1, num_runs + 1), relative_errors, 
         label="Relative Error", marker='x') #plotting relative error on yellow line with x markers

plt.xlabel("Run Number") #labeling the plot
plt.ylabel("Error")
plt.title("Absolute and Relative Error of Monte Carlo Pi (1000 Samples)")
plt.legend()
plt.grid(True)
plt.show()
```
