```python
def is_diagonally_dominant(A):

    # Returns True if A is (row) diagonally dominant, otherwise False.
    # Condition: |a_ii| >= sum_{j!=i} |a_ij| for every row i.

    n = len(A)  # number of rows in the matrix

    # If matrix is empty, it is technically diagonally dominant
    if n == 0:
        return True

    # Check that the matrix is square (same number of rows and columns)
    if any(len(row) != n for row in A):
        raise ValueError("Matrix must be square (n x n).")

    # Loop through each row of the matrix
    for i in range(n):

        # Absolute value of the diagonal element in row i
        diag = abs(A[i][i])

        # Sum of absolute values of all non-diagonal elements in row i
        off_diag_sum = sum(abs(A[i][j]) for j in range(n) if j != i)

        # Check the diagonal dominance condition
        # If diagonal element is smaller than the sum of the others
        if diag < off_diag_sum:
            return False  # matrix is not diagonally dominant

    # If all rows satisfy the condition
    return True


def display_matrix(A):
    for row in A:  # go through each row of the matrix
        # Print each value with formatting so the matrix lines up nicely
        print(" ".join(f"{val:8.3f}" if isinstance(val, float) else f"{val:8}" for val in row))


# Example usage
A = [
    [4, 1, 1],
    [2, 7, 1],
    [1, 1, 5]
]

display_matrix(A)  # print the matrix
print(f"Is the matrix diagonally dominant? {is_diagonally_dominant(A)}")  # check dominance
```
