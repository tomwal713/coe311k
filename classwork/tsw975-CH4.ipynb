```python
import numpy as np

#Number of terms
N = 5000

pi_true = np.pi
partial_sum = 0.0

for k in range(N):
    term = (-1)**k / (2*k + 1) #Leibnitz formula
    partial_sum += term #appending term
    pi_estimate = 4 * partial_sum

    #Print every 100 terms
    if (k + 1) % 100 == 0:
        relative_error = abs(pi_estimate - pi_true) / pi_true
        print(f"Terms: {k+1:4d} | pi estimate: {pi_estimate:.10f} | Relative error: {relative_error:.6e}")
```
