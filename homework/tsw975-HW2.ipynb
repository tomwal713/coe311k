```python
import random #random for simulation
import math #for true value of pi
import numpy as np
import matplotlib.pyplot as plt
```

```python
def monte_carlo_pi(n): #function using to approximate pi
    inside = 0 #counts how many points will fall inside the quarter circle

    for _ in range(n): #here I'm generating points inside the unit square, generating n points
        x = random.random()
        y = random.random()

        if x*x + y*y <= 1: #checking if the points lie within the circle
            inside += 1 #if the point is inside the circle, +1

    return 4 * inside / n #this estimates the area of the quarter circle, multiply by 4 to get pi
```

```python
pi_atan = 4 * math.atan(1) #computed
pi_numpy = np.pi #absolute
sample_sizes = [10, 100, 1000, 10000] #amount of times to run
all_errors = {}
mean_errors = {}
```

```python
for n in sample_sizes:
    errors = []

    for _ in range(10):
        pi_estimate = monte_carlo_pi(n)

        # error compared to numpy.pi (absolute error)
        error = abs(pi_estimate - pi_numpy)
        errors.append(error)

    all_errors[n] = errors #storing each error
    mean_errors[n] = sum(errors) / len(errors) #computing mean error
```

```python
plt.figure()

for n in sample_sizes:
    plt.plot(range(1, 11), all_errors[n], marker='o', label=f"N = {n}")

plt.xlabel("Run Number")
plt.ylabel("Absolute Error")
plt.title("Monte Carlo Pi Error Across Multiple Runs")
plt.legend()
plt.grid(True)
plt.show()
```
