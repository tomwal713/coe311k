```python
import numpy as np
import time
import matplotlib.pyplot as plt


def gauss_seidel(A, b, x0, tol, max_iterations):
    # A: coefficient matrix
    # b: right-hand-side vector
    # x0: initial guess vector
    # tol: stopping tolerance
    # max_iterations: safety cap on iterations

    n = len(b) # Number of unknowns
    x = x0.copy() # Current iterate (starting from initial guess)
    errors = [] # List to store error at each iteration

    # Main iteration loop
    for k in range(max_iterations):

        x_new = x.copy() # Copy current guess

        # Loop over each equation / unknown
        for i in range(n):

            # s1 = sum of terms using the NEWLY updated values (j < i)
            s1 = sum(A[i][j] * x_new[j] for j in range(i))

            # s2 = sum of terms using OLD values (j > i)
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n))

            # Solve equation i for x_i
            x_new[i] = (b[i] - s1 - s2) / A[i][i]

        # If any values blow up to inf or NaN, declare divergence
        if not np.all(np.isfinite(x_new)):
            return x_new, k, False, errors

        # Compute infinity-norm of difference between successive iterates
        # This measures how much the solution changed this step
        err = np.linalg.norm(x_new - x, ord=np.inf)
        errors.append(err)

        # If change is smaller than tolerance, we stop
        if err < tol:
            return x_new, k, True, errors

        # Otherwise continue iteration
        x = x_new

    # If we exit loop, we hit max_iterations without converging
    return x, max_iterations, False, errors


def jacobi(A, b, x0, tol, max_iterations):
    # Same inputs as Gauss-Seidel

    n = len(b) # Number of unknowns
    x = x0.copy() # Current iterate
    errors = [] # Store iteration error

    # Main iteration loop
    for k in range(max_iterations):

        x_new = x.copy() # Jacobi computes entirely from OLD x

        for i in range(n):

            # Sum all terms except the diagonal (j != i)
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)

            # Solve equation i for x_i
            x_new[i] = (b[i] - s) / A[i][i]

        # Check for overflow / divergence
        if not np.all(np.isfinite(x_new)):
            return x_new, k, False, errors

        # Compute infinity-norm error between iterations
        err = np.linalg.norm(x_new - x, ord=np.inf)
        errors.append(err)

        # Stop if tolerance satisfied
        if err < tol:
            return x_new, k, True, errors

        # Move to next iteration
        x = x_new

    return x, max_iterations, False, errors


# Coefficient matrix A
A = np.array([[3.0,  -0.1, -0.2],
              [0.1,   7.0, -0.3],
              [0.3,  -0.2, 10.0]], dtype=float)

# Right-hand-side vector b
b = np.array([7.85, -19.3, 71.4], dtype=float)

# Initial guess x0 = [0, 0, 0]
x0 = np.array([0.0, 0.0, 0.0], dtype=float)

# Tolerances to test
tols = [1e-3, 1e-6, 1e-9]

# Large max iteration limit so it doesn't stop prematurely
max_iterations = 100000


# Running both methods for each tolerance
for tol in tols:

    # Jacobi
    t0 = time.perf_counter() # Start timing
    x_j, k_j, conv_j, err_j = jacobi(A, b, x0, tol, max_iterations)
    t1 = time.perf_counter() # End timing

    # Gauss-Seidel
    t0_gs = time.perf_counter()
    x_gs, k_gs, conv_gs, err_gs = gauss_seidel(A, b, x0, tol, max_iterations)
    t1_gs = time.perf_counter()

    # Printing everything for readability
    print(f"\n=== tolerance: {tol:.0e} ===")

    print("Jacobi:")
    print(f"  Converged: {conv_j}")
    print(f"  Iterations: {k_j + 1 if conv_j else 'N/A'}")
    print(f"  Time (s): {t1 - t0}")
    print(f"  Solution: {x_j}")

    print("Gauss-Seidel:")
    print(f"  Converged: {conv_gs}")
    print(f"  Iterations: {k_gs + 1 if conv_gs else 'N/A'}")
    print(f"  Time (s): {t1_gs - t0_gs}")
    print(f"  Solution: {x_gs}")

    # Plotting convergence behavior

    plt.figure() # Create new figure

    # Plot Jacobi error curve (log scale because it's more clear)
    if len(err_j) > 0:
        plt.semilogy(err_j, label="Jacobi")

    # Plot Gauss-Seidel error curve
    if len(err_gs) > 0:
        plt.semilogy(err_gs, label="Gauss-Seidel")

    plt.xlabel("Iteration")
    plt.ylabel(r"$||x^{(k+1)} - x^{(k)}||_\infty$")
    plt.title(f"Convergence (tol = {tol:.0e})")
    plt.grid(True)
    plt.legend()


# Show all plots
plt.show()

print("\nDiscussion: Smaller tolerances like E-9 vs E-3 means you require a smaller \nchange between iterations, so it generally takes more iterations.")
print("    GaussSeidel usually converges in fewer iterations than Jacobi because \n    it immediately uses the newest updated values.")

# Direct solution (for reference / comparison)
x_direct = np.linalg.solve(A, b)
print("\nDirect solve (reference):")
print(x_direct)
```
