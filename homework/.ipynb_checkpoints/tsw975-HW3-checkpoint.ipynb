```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
# Taylor series approximation function for sin(x)
# sin(x) = sum{n=0}^{âˆž} [(-1)^n * x^(2n+1)] / (2n+1)!

def taylor_sin(x, n_terms):
    """
    Approximate sin(x) using the first n_terms of the Maclaurin series.
    n_terms=1 uses: x
    n_terms=2 uses: x - x^3/3!
    n_terms=3 uses: x - x^3/3! + x^5/5!
    etc.
    """
    x = np.asarray(x, dtype=float)
    s = np.zeros_like(x)
    for n in range(n_terms):
        power = 2*n + 1
        s += ((-1)**n) * (x**power) / math.factorial(power)
    return s
```

```python
# Build x-values and compute approximations (1 to 4 terms)
x = np.arange(0, 10.0 + 0.1, 0.1)  # includes 10.0
exact = np.sin(x)

approxs = {}
for n_terms in [1, 2, 3, 4]:
    approxs[n_terms] = taylor_sin(x, n_terms)

# check at a few points
for test_x in [0, 0.5, 1.0]:
    print(test_x, np.sin(test_x), taylor_sin(test_x, 4))
```

```python
# Absolute errors for each truncation level
errors = {}
for n_terms in [1, 2, 3, 4]:
    errors[n_terms] = np.abs(approxs[n_terms] - exact)

# show max error over the whole interval for each truncation
for n_terms in [1, 2, 3, 4]:
    print(f"{n_terms} term(s): max abs error = {errors[n_terms].max():.6g}")
```

```python
# Put results into a DataFrame (easy to inspect / use for discussion)
df = pd.DataFrame({
    "x": x,
    "sin_exact": exact,
    "sin_1_term": approxs[1],
    "sin_2_terms": approxs[2],
    "sin_3_terms": approxs[3],
    "sin_4_terms": approxs[4],
    "abs_err_1_term": errors[1],
    "abs_err_2_terms": errors[2],
    "abs_err_3_terms": errors[3],
    "abs_err_4_terms": errors[4],
})

df.head(10)
```

```python
# Plotting sin(x) approximations vs exact (optional but cool)
plt.figure()
plt.plot(x, exact, label="exact: sin(x)")
plt.plot(x, approxs[1], label="1 term")
plt.plot(x, approxs[2], label="2 terms")
plt.plot(x, approxs[3], label="3 terms")
plt.plot(x, approxs[4], label="4 terms")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("sin(x): Exact vs Taylor (Maclaurin) truncations")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot truncation (absolute) error vs x for 1â€“4 terms
plt.figure()
plt.plot(x, errors[1], label="|error|, 1 term")
plt.plot(x, errors[2], label="|error|, 2 terms")
plt.plot(x, errors[3], label="|error|, 3 terms")
plt.plot(x, errors[4], label="|error|, 4 terms")
plt.xlabel("x")
plt.ylabel("absolute error")
plt.title("Truncation error vs x (1â€“4 Taylor terms)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Error vs number of terms for a few selected x values
# This directly answers "error as a function of number of terms" for chosen x's
x_samples = [0.5, 1.0, 2.0, 5.0, 10.0]
term_counts = np.arange(1, 5)

plt.figure()
for xv in x_samples:
    exact_val = np.sin(xv)
    err_list = [abs(taylor_sin(xv, n) - exact_val) for n in term_counts]
    plt.plot(term_counts, err_list, marker="o", label=f"x={xv}")
plt.xlabel("number of terms")
plt.ylabel("absolute error")
plt.title("Truncation error vs number of terms (selected x values)")
plt.xticks(term_counts)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Discussion:
# - As I add more terms, the truncation error pretty much decreases, primarily for small x near the origin.
# - It does not decrease uniformly for all x from 0 to 10 with only 4 terms. 
#     For larger x, powers like x^7 and so on get very steep, so truncating early leaves a big remainder.
# - For small x, 2-4 terms is pretty accurate. For larger x, I would need a lot more terms in order to accurately estimate sinx.
```
