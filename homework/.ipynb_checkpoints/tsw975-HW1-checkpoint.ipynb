```python
import random #need random for the simulation
import math #for the true value of pi
import matplotlib.pyplot as plt
```

```python
def monte_carlo_pi(n): #function using to approximate pi
    inside = 0 #counts how many points will fall inside the quarter circle

    for _ in range(n): #here I'm generating points inside the unit square, generating n points
        x = random.random()
        y = random.random()

        if x*x + y*y <= 1: #checking if the points lie within the circle
            inside += 1 #if the point is inside the circle, +1

    return 4 * inside / n #this estimates the area of the quarter circle, multiply by 4 to get pi
```

```python
samples = [] #stores the number of random points the functions used
errors = [] #stores the error in each approximation

for n in range(10, 1001, 20): #looping 
    pi_estimate = monte_carlo_pi(n)
    error = abs(pi_estimate - math.pi) #estimating the error using true values of pi from the imported math

    samples.append(n) #stores samples and error 
    errors.append(error)
```

```python
plt.figure() #now just need to plot the values I stored
plt.plot(samples, errors)
plt.xlabel("Number of Samples")
plt.ylabel("Absolute Error")
plt.title("Monte Carlo Approximation Error for Pi")
plt.grid(True)
plt.show()
```
